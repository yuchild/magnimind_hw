## Step 1: import, check, train, and test

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# from sklearn.datasets import load_boston
# boston = load_boston()
```

```python
"""
ImportError: 
`load_boston` has been removed from scikit-learn since version 1.2.
"""

data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\\s+", skiprows=22, header=None)
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]
```

```python
"""

 The Boston house-price data of Harrison, D. and Rubinfeld, D.L. 'Hedonic
 prices and the demand for clean air', J. Environ. Economics & Management,
 vol.5, 81-102, 1978.   Used in Belsley, Kuh & Welsch, 'Regression diagnostics
 ...', Wiley, 1980.   N.B. Various transformations are used in the table on
 pages 244-261 of the latter.

 Variables in order:
 CRIM     per capita crime rate by town
 ZN       proportion of residential land zoned for lots over 25,000 sq.ft.
 INDUS    proportion of non-retail business acres per town
 CHAS     Charles River dummy variable (= 1 if tract bounds river; 0 otherwise)
 NOX      nitric oxides concentration (parts per 10 million)
 RM       average number of rooms per dwelling
 AGE      proportion of owner-occupied units built prior to 1940
 DIS      weighted distances to five Boston employment centres
 RAD      index of accessibility to radial highways
 TAX      full-value property-tax rate per $10,000
 PTRATIO  pupil-teacher ratio by town
 B        1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town
 LSTAT    % lower status of the population
 MEDV     Median value of owner-occupied homes in $1000's

"""

data[:5]
```

```python
target[:5]
```

```python
cols = ['CRIM','ZN',]
```

```python
# Column names for the feature columns
columns = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']

# Create a DataFrame for the features
df_features = pd.DataFrame(data, columns=columns)

# Add the target column (MEDV) to the DataFrame
df_features['MEDV'] = target

df = df_features.copy()
df
```

## Step 2: Create a scatter plot of each attribute with the mean house price.



```python
# Assuming df_features is the DataFrame with the attributes and target column (MEDV)
attributes = df.columns[:-1]  # Exclude the 'MEDV' column from the attribute list

# Loop through each attribute and create a scatter plot
for attribute in attributes:
    plt.figure(figsize=(6, 4))
    plt.scatter(df_features[attribute], df_features['MEDV'], alpha=0.5)
    plt.title(f'Scatter Plot of {attribute} vs MEDV')
    plt.xlabel(attribute)
    plt.ylabel('MEDV')
    plt.show()
```

## Step 3: Create a box plot of all attributes.

```python
# Create a box plot for each attribute
plt.figure(figsize=(15, 10))  # Adjust the figure size for better readability
df[attributes].boxplot()

# Set title and axis labels
plt.title('Box Plot of Attributes')
plt.xticks(rotation=45)  # Rotate the x-axis labels for better readability
plt.ylabel('Values')
plt.show()
```

## Step 4: Use StandardScaler() to scale the training data set.

```python
from sklearn.preprocessing import StandardScaler

# Assuming df is the original DataFrame with features and target
attributes = df.columns[:-1]  # All columns except the target
target_col = 'MEDV'

# Initialize the StandardScaler
scaler = StandardScaler()

# Scale the features
scaled_features = pd.DataFrame(scaler.fit_transform(df[attributes]), columns=attributes)

# Concatenate the scaled features with the target column
df_scaled = pd.concat([scaled_features, df[target_col]], axis=1)

# Display the scaled DataFrame
df_scaled.head()
```

 ## Step 5: Use KNeighborsRegressor() to fit both not-scaled and scaled datasets. Check the scores of test dataset for both models.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error

# Assuming df is your DataFrame with features and target
attributes = df.columns[:-1]  # All columns except the target (last column is MEDV)
target_col = 'MEDV'

# Step 1: Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df[attributes], df[target_col], test_size=0.2, random_state=42)

# Step 2: Train KNeighborsRegressor on non-scaled data
knn_non_scaled = KNeighborsRegressor()
knn_non_scaled.fit(X_train, y_train)

# Test the model on non-scaled test data
y_pred_non_scaled = knn_non_scaled.predict(X_test)
score_non_scaled = knn_non_scaled.score(X_test, y_test)
mse_non_scaled = mean_squared_error(y_test, y_pred_non_scaled)

# Step 3: Scale the data
scaler = StandardScaler()

# Scale the training and test sets
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Step 4: Train KNeighborsRegressor on scaled data
knn_scaled = KNeighborsRegressor()
knn_scaled.fit(X_train_scaled, y_train)

# Test the model on scaled test data
y_pred_scaled = knn_scaled.predict(X_test_scaled)
score_scaled = knn_scaled.score(X_test_scaled, y_test)
mse_scaled = mean_squared_error(y_test, y_pred_scaled)

# Step 5: Print out the results
print(f"Non-Scaled Data: R^2 score = {score_non_scaled:.4f}, MSE = {mse_non_scaled:.4f}")
print(f"Scaled Data: R^2 score = {score_scaled:.4f}, MSE = {mse_scaled:.4f}")
```

## Step 6: Repeat Step 5 for RandomForestRegressor. Use RandomForestRegressor(n_estimators=100, random_state=0)

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

# Step 2: Train RandomForestRegressor on non-scaled data
rf_non_scaled = RandomForestRegressor(n_estimators=100, random_state=0)
rf_non_scaled.fit(X_train, y_train)

# Test the model on non-scaled test data
y_pred_rf_non_scaled = rf_non_scaled.predict(X_test)
score_rf_non_scaled = rf_non_scaled.score(X_test, y_test)
mse_rf_non_scaled = mean_squared_error(y_test, y_pred_rf_non_scaled)

# Step 4: Train RandomForestRegressor on scaled data
rf_scaled = RandomForestRegressor(n_estimators=100, random_state=0)
rf_scaled.fit(X_train_scaled, y_train)

# Test the model on scaled test data
y_pred_rf_scaled = rf_scaled.predict(X_test_scaled)
score_rf_scaled = rf_scaled.score(X_test_scaled, y_test)
mse_rf_scaled = mean_squared_error(y_test, y_pred_rf_scaled)

# Step 5: Print out the results for RandomForestRegressor
print(f"RandomForest Non-Scaled Data: R^2 score = {score_rf_non_scaled:.4f}, MSE = {mse_rf_non_scaled:.4f}")
print(f"RandomForest Scaled Data: R^2 score = {score_rf_scaled:.4f}, MSE = {mse_rf_scaled:.4f}")
```
