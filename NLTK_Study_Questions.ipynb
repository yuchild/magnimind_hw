```python
"""
Find all the four-letter words in the Chat Corpus (text5). With the help of a frequency distribution (FreqDist), show these words in decreasing order of frequency. 
"""
import nltk
from nltk.corpus import nps_chat
from nltk.probability import FreqDist

# Ensure you've downloaded the necessary corpus
nltk.download('nps_chat')

# Load text5 from the nps_chat corpus
text5 = nps_chat.words()

# Find all four-letter words
four_letter_words = [word.lower() for word in text5 if len(word) == 4]

# Create a frequency distribution for the four-letter words
freq_dist = FreqDist(four_letter_words)

# Display the words in decreasing order of frequency
for word, frequency in freq_dist.most_common():
    print(f"{word}: {frequency}")
```

```python
"""
Define sent to be the list of words ['she', 'sells', 'sea', 'shells', 'by', 'the', 'sea', 'shore']. Now write code to perform the following tasks: 
a. Print all words beginning with sh. 
b. Print all words longer than four characters 

"""
# Define the list of words
sent = ['she', 'sells', 'sea', 'shells', 'by', 'the', 'sea', 'shore']

# a. Print all words beginning with 'sh'
words_starting_with_sh = [word for word in sent if word.startswith('sh')]
print("Words beginning with 'sh':", words_starting_with_sh)

# b. Print all words longer than four characters
words_longer_than_four = [word for word in sent if len(word) > 4]
print("Words longer than four characters:", words_longer_than_four)
```

```python
"""
What does the following Python code do? sum([len(w) for w in text1]) Can you use it to work out the average word length of a text? 
"""
total_characters = sum([len(w) for w in text5])
total_words = len(text5)
average_word_length = total_characters / total_words
print("Average word length:", average_word_length)
```

```python
"""
Define a function called vocab_size(text) that has a single parameter for the text, and which returns the vocabulary size of the text. 
"""
def vocab_size(text):
    return len(set(text))

text = ["she", "sells", "sea", "shells", "by", "the", "sea", "shore"]
print(vocab_size(text))  # Output: 7 (since "sea" is repeated)
```

```python
"""
Define a function percent(word, text) that calculates how often a given word occurs in a text and expresses the result as a percentage. 
"""
def percent(word, text):
    word_count = text.count(word)  # Count occurrences of the word
    total_words = len(text)        # Total number of words in the text
    return (word_count / total_words) * 100  # Calculate percentage

text = ["she", "sells", "sea", "shells", "by", "the", "sea", "shore"]
print(percent("sea", text))  # Output: 25.0 (since "sea" appears twice in 8 words)

```
